### 1.1 Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()

```python
def myreduce(func,seq):
    result=seq[0]
    for i in range(1,len(seq)):
        result=func(result,seq[i])
    return result
```

```python
find_max=lambda a,b:a if (a>b) else b  
a=myreduce(find_max,[1,2,3,4,5])
print(a)
```

```python
b=myreduce(lambda a,b:a+b,[2,3,4,8])
print(b)
```

```python
def sum1(a,b):
    return a+b
d=myreduce(sum1,(10,20,30,40))
print(d)
```

### 1.2 Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()

```python
def myfilter(func,seq):
    result=[]
    for i in seq:
        if func(i):
            result.append(i)
    return result
```

```python
def iseven(n):
    return n%2==0
p=myfilter(iseven,[1,2,3,4,5])
```

```python
print(p)
```

```python
q=myfilter(lambda a:1 if a%2==0 else 0,[10,20,15,25,30,35])
print(q)
```

### 3. Implement a function longestWord() that takes a list of words and returns the longest one.

```python
def longestWord(lst):
    counts={}
    for i in lst:
        counts[i]=len(i)
    m=max(counts,key=counts.get)
    return m
```

```python
l=longestWord(['rutuja','jasmine','ruhi'])
print("Longest word is : %s"%(l))
```

### 1.1 Write a Python Program(with class concepts) to find the area of the triangle using the below formula.
area = (s*(s-a)*(s-b)*(s-c)) *** 0.5
Function to take the length of the sides of triangle from user should be defined in the parent
class and function to calculate the area should be defined in subclass.

```python
class Parent:
    def take_input(self):
        self.a=int(input("Enter First Side: "))
        self.b=int(input("Enter Second Side: "))
        self.c=int(input("Enter Third Side: "))
class Child(Parent):
    def calculate_area(self):
        s=(self.a+self.b+self.c)/2
        area=(s*(s-self.a)*(s-self.b)*(s-self.c))**0.5
        print(area)
```

```python
c=Child()
c.take_input()
c.calculate_area()
```

### 1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n.

```python
def filter_long_words(words,n):
    longer_words=[]
    for i in words:
        if len(i)>n:
            longer_words.append(i)
    return longer_words
```

```python
filter_long_words(['Dinesh','Neha','Ram','Riya','Tina','Daksh','Tanmay'],4)
```

### 2.1 Write a Python program using function concept that maps list of words into a list of integers representing the lengths of the corresponding words.
Hint: If a list [ ab,cde,erty] is passed on to the python function output should come as [2,3,4] Here 2,3 and 4 are the lengths of the words in the list.

```python
def map_lengths(words):
    lengths=[]
    for i in words:
        lengths.append(len(i))
    return lengths
```

```python
map_lengths(['ab','cde','erty'])
```

```python
map_lengths(['Rabbit','Lion','Hen'])
```

### 2.2 Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is a vowel, False otherwise.

```python
def is_vowel(char):
    vowels=('a','e','i','o','u')
    if char in vowels:
        return True
    else :
        return False
```

```python
print(is_vowel('a'))
print(is_vowel('z'))
```
